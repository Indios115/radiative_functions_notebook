**SOLAR AND TERRESTRIAL SPECTRUM**


```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
T_sun = 6000 #Kelvin
```

```python
T_earth = 290  #Kelvin
```

```python
h = 6.6260687*10**-34 # Planck constant in Js
c = 2.99792458*10**8; # photon speed in m/s
k = 1.3806503*10**-23; # Boltzmann constant in J/K

  # Costanti della legge di Planck
c1 = 2*h*c**2  # [W*m^2] 
c2 = h*c/k   # [m*K]
```

```python
lambda_1 = np.arange(0.1,100,0.01) # micron
lambda_m = lambda_1*10**-6 #m
```

```python
import numpy as np

def B_lambda(T, lambda_1):
    """
   Compute spectral radiance of a blackbody (as a function of wavelenght).

    Parametri:
    ----------
    T : Temperature (Kelvin)
    lambda_1 : Wavelenght(micron)

    Ritorna:
    --------
    B : blackbody radiance
    
    """
    
  
    # Assicurati che lambda_1 sia un array NumPy
    lambda_1 = np.asarray(lambda_1)
    
    # Calcolo vettorializzato
    B = c1 / ((lambda_1**5) * (np.exp(c2 / (lambda_m* T)) - 1))  # W/m2/sr/m
    
    return B
```

```python

# Plot
plt.figure(figsize=(8, 6))

# Radianza per le temperature
plt.plot(lambda_1, 10**-6*B_lambda(T_earth, lambda_m), label="T = {} K (Earth)".format(T_earth), color='blue')
plt.plot(lambda_1, 10**-12*B_lambda(T_sun, lambda_m), label="T = {} K (Sun) scaled 10**-6".format(T_sun), color='yellow')

# Etichette degli assi
plt.xlabel("Wavelength (um)", fontsize=12)
plt.ylabel("Radiance (W/m/sr/um)", fontsize=12)


plt.xscale("log")

# Titolo e legenda
plt.legend(loc="upper right", fontsize=10)

# Griglia
plt.grid(True, linestyle='--', alpha=0.7)
```

**PLOTTING PLANK (WIEN LAWS)**

```python
wn = np.arange(0.01,3000,0.0001) # cm-1
```

```python
wn_m = wn*100
```

```python
def B_wn(T, wn):
    """
   Compute spectral radiance of a blackbody (as a function of wavenumber).

    Parametri:
    ----------
    T : Temperature (Kelvin)
    lambda_1 : Wavelenght(micron)

    Ritorna:
    --------
    B : blackbody radiance
    
    """
    
  
    # Assicurati che lambda_1 sia un array NumPy
    wn = np.asarray(wn)
    
    # Calcolo vettorializzato
    B = c1 *wn**3 / ((np.exp(c2 *wn / (T)) - 1))  # W/m2/sr/m-1
    
    return B
```

```python
T1 = 255 #Kelvin
T2 = 280
T3 = 350
```

```python
plt.figure(figsize=(8, 6))


plt.plot(wn, 10**5*B_wn(T2,wn_m), label="T {}".format(T2), color='red')  # mW/m2/sr/cm-1
plt.plot(wn, 10**5*B_wn(T3, wn_m), label="T {}".format(T3), color='violet') # mW/m2/sr/cm-1

# Etichette degli assi
plt.xlabel("Wavenumber (cm)", fontsize=12)
plt.ylabel("Radiance (W/m/sr/cm)", fontsize=12)

# Titolo e legenda
plt.title("Radiance vs Wavenumber", fontsize=14)
plt.legend(loc="upper right", fontsize=10)

# Griglia
plt.grid(True, linestyle='--', alpha=0.7)
```
