Wien's displacement Law

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
h = 6.6260687*10**-34 # Planck constant in Js
c = 2.99792458*10**8; # photon speed in m/s
k = 1.3806503*10**-23; # Boltzmann constant in J/K

  # Costanti della legge di Planck
c1 = 2*h*c**2  # [W*m^2] 
c2 = h*c/k   # [m*K]
```

```python
import numpy as np

def B_lambda(T, lambda_m):
    """
   Compute spectral radiance of a blackbody (as a function of wavelenght).

    Parametri:
    ----------
    T : Temperature (Kelvin)
    lambda_m : Wavelenght(micron)

    Ritorna:
    --------
    B : blackbody radiance
    
    """
    

    # Calcolo
    B = c1 / ((lambda_m**5) * (np.exp(c2 / (lambda_m* T)) - 1))  # W/m2/sr/m
    
    return B
```

```python
T_1 = 3000
T_2 = 3500
T_3 = 4000
T_4 = 5000
lambda_um = np.arange(0.1,5.,0.00001)
lambda_m = lambda_um*10**-6 #m
```

```python
# Plot
plt.figure(figsize=(8, 6))

# Radianza per le temperature (10**-6 B da m a um)
plt.plot(lambda_um, 10**-6* B_lambda(T_1, lambda_m), label=  " T = {} K ".format(T_1), color='blue')
plt.plot(lambda_um, 10**-6* B_lambda(T_2, lambda_m), label=  " T = {} K ".format(T_2), color='orange')
plt.plot(lambda_um, 10**-6* B_lambda(T_3, lambda_m), label=  " T = {} K ".format(T_3), color='red')
plt.plot(lambda_um, 10**-6* B_lambda(T_4, lambda_m), label=  " T = {} K ".format(T_4), color='violet')

# Etichette degli assi
plt.xlabel("Wavelength (um)", fontsize=12)
plt.ylabel("Radiance (W/m²/sr/um)", fontsize=12)


#plt.xscale("log")

# Titolo e legenda
plt.legend(loc="upper right", fontsize=10)

# Griglia
plt.grid(True, linestyle='--', alpha=0.7)
```
